```python
import pandas as pd
import random
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.DataFrame()
```

```python
from re import X
x = pd.Series([12,25,68,42,113])
y = pd.Series([11,29,58,121,100])
```

```python
df['x'] = x
df['y'] = y
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,3))

#plotting scatterplots on each axes

ax1.scatter(df['x'], df['y'], color = 'purple')
ax2.scatter(df['x'] * 2, df['y'] * 2, color = 'red')

ax1.set_title("Covariance - " + str(np.cov(df['x'],df['y'])[0,1]))

ax2.set_title("Covariance - " + str(np.cov(df['x']*2,df['y']*2)[0,1]))
```
