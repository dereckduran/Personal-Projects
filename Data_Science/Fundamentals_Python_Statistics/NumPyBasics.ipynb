```python
import pandas as pd

# Sample data for the first DataFrame
data1 = {
    'user_id': [1, 2, 3, 4],
    'user_name': ['Alice', 'Bob', 'Charlie', 'David']
}

# Sample data for the second DataFrame
data2 = {
    'user_id': [3, 4, 5, 6],
    'song_id': [101, 102, 103, 104],
    'duration': [300, 200, 400, 600]
}

# Creating DataFrames
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Inner join (only matching keys)
inner_join = pd.merge(df1, df2, on='user_id', how='inner')

# Outer join (all keys from both DataFrames)
outer_join = pd.merge(df1, df2, on='user_id', how='outer')

# Left join (all keys from the left DataFrame)
left_join = pd.merge(df1, df2, on='user_id', how='left')

# Right join (all keys from the right DataFrame)
right_join = pd.merge(df1, df2, on='user_id', how='right')

# Display the results
print("Inner Join:")
print(inner_join)
print("\nOuter Join:")
print(outer_join)
print("\nLeft Join:")
print(left_join)
print("\nRight Join:")
print(right_join)
```

```python
def twodigit(num):
    hash = {}

    for num in str(num):
        hash[num] = hash.get(num, 0) + 1

    return len(hash) == 2
```

```python
print(twodigit(num))
```

```python
profits = [-1, 9, 0, 8, -5, 6, -24]
```

```python
def max_profit(numbers):
    left, right = 0, len(numbers) - 1
    maxsum = 0
    while left < right:
        sum = 0
        for num in range(left, right):
            sum += numbers[num]

        if sum > maxsum:
            maxsum = sum

        left += 1
        right -= 1

    return maxsum
```

```python
print(max_profit(profits))
```
